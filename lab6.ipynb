```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
sql = " select * from income inner join name on name.fips = income.fips "
```

```python
import pandas as pd

df = pd.read_sql(sql, conn)
df.head()
```

```python
import matplotlib.pyplot as plt

# Sort the dataframe by state and year
df_sorted = df.sort_values(['name', 'year'])

# Calculate the year-over-year income growth rate
df_sorted['income_growth_rate'] = df_sorted.groupby('name')['income'].pct_change()

# Pivot the dataframe to create a dataframe with years as index and states as columns
df_pivot = df_sorted.pivot(index='year', columns='name', values='income_growth_rate')

# Plot the year-over-year income growth rate for each state
plt.figure(figsize=(10, 8))
for column in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Year-over-year income growth rate for each state", loc='left', fontsize=12, fontweight=0)
plt.xlabel("Year")
plt.ylabel("Income growth rate")
plt.show()
```
